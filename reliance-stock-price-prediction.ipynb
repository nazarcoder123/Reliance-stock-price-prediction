```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.linear_model import LinearRegression,LogisticRegression
from sklearn.metrics import accuracy_score,r2_score
from sklearn.preprocessing import StandardScaler,MinMaxScaler
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv("/kaggle/input/reliance-nse-stock-data/RILO - Copy.csv")
```

```python
df.sample(3)
```

```python
x = df[['Prev Close','Open Price','High Price','Low Price','Close Price','Deliverable Qty','% Dly Qt to Traded Qty']]
x
```

```python
y = df.iloc[:,9]
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.head()
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.predict(x_test)
```

```python
y_pred = lr.predict(x_test)
```

```python
r2_score(y_pred,y_test)
```

```python
cross_val_score(lr,x_train,y_train,scoring='r2').mean()
```
